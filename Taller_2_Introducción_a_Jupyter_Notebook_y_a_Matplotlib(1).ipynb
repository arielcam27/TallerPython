## LaTeX

Por otro lado, $\LaTeX$ es un lenguaje que nos permite, entre otras cosas, mostrar símbolos matemáticos y ecuaciones de forma elegante. Para mayor información: https://es.wikipedia.org/wiki/LaTeX

En Jupyter, teniendo $\LaTeX$ instalado en la máquina podemos usar dicho lenguaje en las celdas de Markdown. Esencialmente hay dos formas de visualizar texto generado con $\LaTeX$ en Jupyter: ya sea con `$...$` ó con `$$...$$`. La diferencia entre ambos es que en el primero la salida se mostrará en la misma línea mientras que en el segundo la salida se mostrará en una nueva línea.

`Tenemos que $a+b>0$.`

Tenemos que $a+b>0$.

`Tenemos que $$a+b>0.$$`

|`\cdot`|$\cdot$|

# Funciones en Python

Podemos crear funciones personalizadas en Python usando el siguiente esquema:

```
def miFuncion( argumentos ):
  codigo
  return salida
```

Por ejemplo, podríamos tener la necesidad de usar una función real como $f(x)=e^x-x-1$, y para ello podríamos definir una función con argumento `x`.

````python
import math
````

````python
def miFuncion(x):
  resultado = math.exp(x) - x - 1
  return resultado
````

````python
print( miFuncion(0), miFuncion(1), miFuncion(-1) )
````

````python
def repiteOracion(oracion="Hola", num=1):
  orFinal = oracion * num
  print(orFinal)
````

## Operaciones en listas: Comprensión de listas

En varias ocasiones estamos interesados en transformar a los elementos de una lista mediante una operación o función. ¿Qué pasa si lo intentamos hacer directamente?

````python
lst1 = [1,2,3]
print( lst1*2 ) # si intentamos multiplicar una lista obtendremos la lista repetida
````

```
print( lst1**2 ) # si intentamos elevar al cuadrado una lista obtendremos un error

TypeError                                 Traceback (most recent call last)

<ipython-input-22-3045760dfda2> in <cell line: 1>()
----> 1 print( lst1**2 )

TypeError: unsupported operand type(s) for ** or pow(): 'list' and 'int'
```

La manera adecuada de tratar de hacer esto con listas es con el concepto de **comprensión de listas**. Tiene la siguiente estructura:
```
[ (operacion) for x in ... ]
```

````python
lst1 = [1,2,3]
print( [ 2*x for x in lst1 ] )
````

## NumPy: Arreglos

Un tipo especial de listas son los **arreglos NumPy** ("NumPy arrays"). Son similares a las listas: son mutables ("modificables"). Para usarlos, primero necesitaremos cargar el módulo de NumPy usando `import numpy as np`. Después se utiliza `np.array()`.

````python
# importamos módulo
import numpy as np
````

````python
# creamos arreglo
an1 = np.array([1,2,3,4])
print(an1)
````

Podemos crear arreglos rectangulares similares a matrices. La idea es hacer una *lista de listas*: cada lista interior será un renglón del arreglo.

$$\begin{pmatrix}a_1 & a_2 \\ b_1 & b_2\end{pmatrix}$$

```
np.array([a1, a2],
         [b1, b2]])
```

````python
# "matriz identidad de 2x2"
an2 = np.array([ [1,0],
                 [0,1]])
print(an2)
````

````python
# podemos acceder a elementos de un arreglo rectangular usando dos índices
na3 = np.array([[4,7,-5],[-3,11,0]])

# mostrar entrada (1,2) de an3
print( na3[0,1] )

# mostrar todo el arreglo
print( na3 )
````

Podemos crear arreglos muy útiles usando `np.linspace`: nos permite crear arreglos desde cierto valor "a" hasta un valor "b" y se podrían especificar cuántos números equiespaciados entre esos valores se quieren. Ver documentación: https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

````python
# crear un arreglo con los números del 1 al 10
an3 = np.linspace(1, 10, num=10) # el parámetro "num" dicta el número de elementos que se quieren
print(an3)
````

````python
arr = np.linspace(1, 10) # si no se especifica valor de "num" toma el valor de 50 por default
print(arr.shape) # imprimir tamaño de arreglo
````

Uno similar es `np.arange`, en el cual especificamos el valor inicial, el valor final, y el **salto** entre un elemento y el siguiente del arreglo. **Nota importante:** en `arange` el valor final ya no se incluye, por lo que hay que tener cuidado. Ver documentación: https://numpy.org/doc/stable/reference/generated/numpy.arange.html

````python
# crear un arreglo con los números del 1 al 10
an4 = np.arange(1,11,step=1) # el parámetro "step" dicta cómo "brincar"; notar que se pone 11 en vez de 10
print(an4)
````

````python
an5 = np.arange(1,11) # si no se especifica "step" se tomará un valor de 1 por default
print(an5)
````

## Matplotlib: Gráficas

Para graficar usaremos la paquetería dada por `matplotlib`. Para importar usamos `import matplotlib.pyplot as plt`. Para usar sus funciones usaremos `plt.` seguido de la función.

````python
# importar pyplot de matplotlib y ponerle alias plt
import matplotlib.pyplot as plt
````

Para crear una figura en blanco, declaramos `plt.figure()`. Después, podemos usar funciones que nos permitan "dibujar" en esa figura, como por ejemplo `plt.plot()` ó `plt.scatter()`. Para mostrar la figura final, ejecutamos `plt.show()`.

````python
x = np.arange(0,5)
y = x ** 2 # esto crea una tupla "y" con los elementos de "x" al cuadrado

plt.figure() # figura en blanco
plt.scatter(x, y) # para eje horizontal se usarán los valores en "x", para eje vertical los valores en "y"
plt.show() # mostrar la gráfica
````

````python
x = np.linspace(0,5,num=100) # aquí el arreglo va de 0 a 5, usando 100 elementos en total
y = x ** 2 # esto crea una tupla "y" con los elementos de "x" al cuadrado

plt.figure() # figura en blanco
plt.plot(x, y) # para eje horizontal se usarán los valores en "x", para eje vertical los valores en "y"
plt.show() # mostrar la gráfica
````

## Modificando estilo de gráficas

En Matplotlib es posible modificar la apariencia de diversos elementos, como por ejemplo:
* color de líneas
* estilo de líneas
* marcadores en líneas
* tamaño de fuente
* límite en ejes horizontal y vertical

Este tipo de detalles estéticos pueden resultar de gran importancia para compartir nuestro trabajo con otras personas, ya sean reportes, artículos u otros documentos.

````python
xcor = np.linspace(-1,1)
ycor = xcor**2-1

plt.figure(figsize=(3,3),layout='compressed') # 'figsize' cambia el tamaño de la figura final
plt.plot(xcor,ycor,linestyle='--',color='r',linewidth=3)
# en plot
# linestyle: estilo de línea
# color: color de línea
# linewidth: ancho de línea

plt.scatter([0.5],[0.5**2-1],s=300,marker='*',color='b')
# en scatter
# s: tamaño de los marcadores
# marker: estulo de marcador

# etiquetar los ejes
plt.xlabel("Eje horizontal",size=14)
plt.ylabel("Eje vertical",size=14)
# size: tamaño de fuente

# modificar etiquetas en ejes
plt.xticks(size=12)
plt.yticks(size=12)

plt.grid() # agregar rejilla para referencia visual
plt.show()
````

## Múltiples gráficas con `subplots`

En ocasiones tendremos interés en mostrar más de una gráfica en una misma figura. Esto es posible ya sea poniendo las gráficas en el mismo lienzo o usando subplots. Esto puede hacerse usando `plt.subplot()` ó `plt.subplots()`.

````python
xx=np.linspace(0,1)
y1=xx**2
y2=2*xx

plt.figure(layout='tight', figsize=(6,3))

plt.subplot(1,2,1)
plt.plot(xx,y2,label='lineal')
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.legend()
plt.grid()

plt.subplot(1,2,2)
plt.plot(xx,y1,label='cuadrática')
plt.xlabel("eje x")
plt.ylabel("eje y")
plt.legend()
plt.grid()

plt.show()
````

Para usar "subplots" necesitamos comprender un aspecto técnico sobre su funcionamiento. Al usar `subplots` se generará un objeto `Figure` y otro `Axes` (o arreglo de ellos). La figura es como el lienzo completo sobre el cual pondremos los "axes" que corresponden a cada plot. Las funciones disponibles para "axes" se pueden consultar en la documentación: https://matplotlib.org/stable/api/axes_api.html

````python
fig, axs = plt.subplots(2, 1, figsize=(2, 4))
axs[0].plot(xx,y1,label='cuadrática')
axs[0].legend()
axs[1].plot(xx,y2,label='lineal')
plt.legend()

plt.show()
````

En ocasiones es conveniente que las figuras compartan alguno de los ejes. En este caso, se pueden usar los argumentos `sharey=True` o `sharex=True` en `subplots` para obtener esto.
